```python
import os
import wfdb
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import FastICA, PCA

from scipy.signal import find_peaks, butter, filtfilt
from scipy.stats import pearsonr

import scipy.signal as signal


```

```python
# Paths
data_dir = 'set-a-ext'
ica_output_dir = 'ica_outputs_ext'
os.makedirs(ica_output_dir, exist_ok=True)

# List of record names
records = [f"a{str(i).zfill(2)}" for i in range(1, 76)]
```

```python
# Path for combined signals and ICA output

ica_output_dir = 'ica_outputs2'
os.makedirs(ica_output_dir, exist_ok=True)

subdirectory = [f"sub{str(i).zfill(2)}" for i in range(1, 11)]
levels = [f"l{i}" for i in range(1, 6)]

for record_name in records:
    
    print(f"Processing: {record_name}")
    try:
        # Load combined signals
        record = wfdb.rdrecord(os.path.join(data_dir, record_name))
        mixed = record.p_signal

        # Apply PCA to reduce dimensionality

        pca = PCA(n_components=4, whiten=True, random_state=0)
        pca_output = pca.fit_transform(mixed) 

        # Apply ICA
        ica = FastICA(n_components=4, random_state=0)
        separated = ica.fit_transform(mixed)

        # Save output (as NumPy array)
        np.save(os.path.join(ica_output_dir, f"{record_name}_ica.npy"), separated)

    except Exception as e:
        print(f"Error: {e}")
    
        
```

```python
base_dir = './set-a-ext'
ica_output_dir = 'ica_outputs2'
results_dir = 'ica_analysis_results2'
cleaned_dir = 'ica_cleaned_signals2'
os.makedirs(results_dir, exist_ok=True)
os.makedirs(cleaned_dir, exist_ok=True)


def compute_hr(peaks, fs):
    rr_intervals = np.diff(peaks) / fs
    return 60 / np.mean(rr_intervals) if len(rr_intervals) > 0 else 0

for record_name in records:
    print(f"Analyzing: {record_name}")
    try:
        # Load ICA components from saved numpy file
        ica_path = os.path.join(ica_output_dir, f"{record_name}_ica.npy")
        separated = np.load(ica_path)

        # Load original record to get fs and annotation
        record_path = os.path.join(base_dir, record_name)
        record = wfdb.rdrecord(record_path)
        fs = record.fs

        # Load QRS annotations (assuming fqrs extension)
        ann = wfdb.rdann(record_path, 'fqrs')
        rpeaks = ann.sample
        true_hr = compute_hr(rpeaks, fs)

        # Prepare result file
        result_file = os.path.join(results_dir, f"{record_name}_analysis.txt")
        with open(result_file, 'w') as f:
            f.write(f"=== Analysis: {record_name} ===\n")
            f.write(f"Sampling Frequency: {fs} Hz\n")
            f.write(f"True HR (from annotation): {true_hr:.1f} bpm\n\n")

            correlations = []
            for i in range(separated.shape[1]):
                comp_signal = separated[:, i]
                peaks, _ = find_peaks(comp_signal, distance=fs*0.4)
                est_hr = compute_hr(peaks, fs)

                # Correlate component with original signal (optional)
                corr, _ = pearsonr(comp_signal, record.p_signal[:, 0])  # Assuming channel 0 for comparison

                correlations.append((corr, i))

                f.write(f"--- Component {i+1} ---\n")
                f.write(f"Estimated HR: {est_hr:.1f} bpm\n")
                f.write(f"Correlation with original signal: r = {corr:.3f}\n\n")

            # Select best component by correlation magnitude
            best_idx = max(correlations, key=lambda x: abs(x[0]))[1]
            f.write(f"Selected Component: {best_idx + 1}\n")

        # Save cleaned signal (best component)
        cleaned_signal = separated[:, best_idx]
        np.save(os.path.join(cleaned_dir, f"{record_name}_cleaned.npy"), cleaned_signal)

        print(f"Saved: {result_file}")
        print(f"Saved cleaned component for {record_name}")

    except Exception as e:
        print(f"Error processing {record_name}: {e}")
```

```python

ica_output_dir = 'ica_outputs_2'
record_name = 'a03'

# Load original record
record_path = os.path.join(base_dir, record_name)
record = wfdb.rdrecord(record_path)
original_signal = record.p_signal[:, 0]  # channel 0

# Load ICA components
ica_path = os.path.join(ica_output_dir, f"{record_name}_ica.npy")
separated = np.load(ica_path)

# Reduce number of samples for clearer plot
n_samples = 2000
original_signal_plot = original_signal[:n_samples]
separated_plot = separated[:n_samples, :]

# Plot original signal
plt.figure(figsize=(12, 3))
plt.plot(original_signal_plot)
plt.title(f'Original Signal - {record_name} (first {n_samples} samples)')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()

# Plot independent components in 4 subplots
plt.figure(figsize=(12, 8))
n_components = separated_plot.shape[1]
for i in range(n_components):
    plt.subplot(n_components, 1, i+1)
    plt.plot(separated_plot[:, i])
    plt.title(f'Independent Component {i+1}')
    plt.ylabel('Amplitude')
    plt.grid(True)
plt.xlabel('Samples')
plt.tight_layout()
plt.show()
```
